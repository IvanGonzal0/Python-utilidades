##Machine Learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Vamos a definir los datos y columnas q vamos autilizar para nuestro modelo

X = datos_ml[['COD_UPZ_GRUPO']] #aca tengo un dataframe de pandas
y = datos_ml['Precio_Millon']   #aca tengo una serie de pandas

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state= 99) #random state es para q sea probabilistico
```

```python
X_train, X_test
```

```python
y_train, y_test
```

```python
#instanciamos el modelo

modelo = LinearRegression()
```

```python
#le hacemos un ajuste de datos y lo entrenamos

modelo.fit(X_train, y_train)
```

```python
#prediccion
y_predict_test = modelo.predict(X_test)
```
