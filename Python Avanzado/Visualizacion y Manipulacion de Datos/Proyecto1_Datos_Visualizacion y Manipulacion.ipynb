##Punto 2

2) Para el dataset del ejercicio 1, sumar todas las ventas para cada producto durante el
año. Luego hacer un gráfico de torta mostrando el porcentaje de ventas de cada
categoría respecto al total.


```python
import plotly.express as px
```

```python
totales = df1[['facecream',	'facewash',	'toothpaste',	'bathingsoap',	'shampoo',	'moisturizer']].sum() #del df del ejercicio 1 calculo la suma de todas las categorias
columnas = ['total_ventas'] #voy a crear una columna con el total de esa sumatoria
df_news = pd.DataFrame(totales, columns = columnas) #aca armo el nuevo df para graficar el pie

plt.pie(df_news['total_ventas'], labels=df_news.index, autopct="%0.1f %%")  #Grafico el porcentaje total de las ventas en el año por categoria
plt.title('Porcentaje de ventas por producto por año')
plt.show()
```

```python
#REALIZANDO EL GRAFICO CON PLOTLY, grafico interactivo
fig = px.pie(df_news, values='total_ventas', names=df_news.index, title='Porcentaje de ventas por producto al año',color_discrete_sequence=px.colors.sequential.RdBu) #grafico la suma de todas las categorias(lo hice en el grafico anterior, pero no me gusta como queda, éste grafico es mas bonito)
fig.update_traces( textposition='inside',textinfo='percent+label')      #update_traces me sirve para las apariencias e informacion de los textos en el grafico
fig.update_layout(title_font_family='Arial Black', title_font_size= 30) #update_layout me sirve para modificar apariencias de la capa sobre la que estoy trabajando(fuente, tamaño,color), esto me da permiso para los titulos tambien
fig.show()
```

### b) Crear un sub-set de datos (dataframe) con todos los partidos que se jugaron en la ciudad de Río de Janeiro. Con este, hacer un gráfico que muestre por cada partido (fecha), el total de goles (sumando los goles del equipo local y el visitante).

```python
df_wcMatch_rio = df_wcMatches[df_wcMatches['City'] == 'Rio De Janeiro ']

df1 = df_wcMatch_rio[['Datetime', 'Home Team Name', 'Home Team Goals', 'Away Team Goals', 'Away Team Name']]  #aca extraigo y hago un subdataset de los datos que me interesan para tener mejor vision
df1['TotalGoals'] = df_wcMatch_rio['Home Team Goals'] + df_wcMatch_rio['Away Team Goals'] #aca sumo los goles de los 2 equipos
df1['Encuentro']  = df1['Home Team Name'] + ' - ' + df1['Away Team Name'] #creo una nueva columna que se llama encuentro para guardar los 2 equipos que jugaron esa fecha, para luego hacer un grafico interactivo con plotly
df1.head()
```

```python
#Grafico con plotly
fig = px.bar(df1, x="Datetime", y="TotalGoals", color="Encuentro", title="Paises que ganaron un mundial, y cantidad de goles que hicieron")
fig.update_layout(font_family="Rockwell",font_color="black",
                  title_font_color="lightseagreen",title_font_size= 28,legend_title_font_color="green")
fig.show()
```
